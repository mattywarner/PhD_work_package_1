```python
error_dislo_111 = 0.01*interp_exp_111_stress
error_dislo_001 = 0.01*interp_exp_001_stress
error_pheno_111 = np.linspace(5, 0, len(interp_exp_111_stress)) + 0.01*interp_exp_111_stress
error_pheno_001 = np.linspace(5, 0, len(interp_exp_001_stress)) + 0.01*interp_exp_001_stress
errors = [error_dislo_111, error_dislo_001, error_pheno_111, error_pheno_001]

def custom_loss(params, Y_actual, Y_error, error_scale=1.0):

    params = np.array(params).reshape(1, -1)

    #y_prediction,y_prediction_error = surrogates[k].make_prediction(X=params,return_std=True)

    if k == 0 or k == 1: #Dislotwin GPs

        y_prediction_111,y_prediction_error_111 = surrogates[0].make_prediction(X=params,return_std=True)
        y_prediction_001,y_prediction_error_001 = surrogates[1].make_prediction(X=params,return_std=True)

    elif k == 2 or k == 3: #Phenopowerlaw GPs

        y_prediction_111,y_prediction_error_111 = surrogates[2].make_prediction(X=params,return_std=True)
        y_prediction_001,y_prediction_error_001 = surrogates[3].make_prediction(X=params,return_std=True)

    loss_111 = (y_prediction_111[0]-Y_actual[0])**2/(Y_error[0])**2 + error_scale*y_prediction_error_111**2/(Y_error[0])**2
    loss_001 = (y_prediction_001[0]-Y_actual[1])**2/(Y_error[1])**2 + error_scale*y_prediction_error_001**2/(Y_error[1])**2

    return loss_111.sum() + loss_001.sum()

results = []

for k in range(len(surrogates)):
    #if k == 0 or k == 1:
    #    Y_actual, Y_error = interp_exp_stress[0:2], errors[0:2]
    #if k == 2 or k == 3:
    #    Y_actual, Y_error = interp_exp_stress[2:4], errors[2:4]

    res = surrogates[k].fit(Y_actual = interp_exp_stress[k],
                    Y_error = errors[k],
                    iters = 5,
                    n = 250,
                    use_std = True,
                    #error_scale=50
                    #loss_func = custom_loss
                    )
    results.append(res)

    print(res)
```

# 4. Bayesian Inference from MCMC sampling

```python
inference_data_list = []
for m in range(len(surrogates)):
    #if m == 0 or m == 1:
    #    Y_actual, Y_error = interp_exp_stress[0:2], errors[0:2]
    #if m == 2 or m == 3:
    #    Y_actual, Y_error = interp_exp_stress[2:4], errors[2:4]

    inference_data = surrogates[m].perfom_inference(Y_actual=interp_exp_stress[m],
                                            Y_error=errors[m],
                                            draws=50_000,
                                            tune=50_000,
                                            chains=4,
                                            cores=4,
                                            return_inferencedata=True,
                                            initval=results[m].x,
                                            use_std=True,
                                            #error_scale=50
                                            )
    inference_data_list.append(inference_data)
```

```python
for n in range(len(inference_data_list)):
    print(arviz.summary(inference_data_list[n]))
```

```python
for o in range(len(inference_data_list)):
    axes = arviz.plot_trace(
        data=inference_data_list[o],
        compact=True,
        backend_kwargs={"figsize": (12, 9), "layout": "constrained"},
    )
    plt.gcf().suptitle("Trace", fontsize=16)
    plt.gcf().set_figwidth(5)
    plt.gcf().set_figheight(5)
```

# Posterior extraction

```python
posterior_info = []
for p in range(len(inference_data_list)):
    X_posterior = np.array([np.array(inference_data_list[p].posterior[x]).flatten() for x in surrogates[p].parameter_names]).T
    X_mode = np.array([scipy.stats.mode(X_posterior[:,i])[0] for i in range(len(surrogates[p].parameter_names))])
    X_mean = np.array([np.mean(X_posterior[:,i]) for i in range(len(surrogates[p].parameter_names))])
    X_std = np.array([np.std(X_posterior[:,i]) for i in range(len(surrogates[p].parameter_names))])
    posterior_info.append([X_posterior, X_mode, X_mean, X_std])
```
