```python
import pymc as pm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Times New Roman'
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1)
ax.hlines(0.5, 1, 40, color = 'r')


x = np.linspace(1,40, 1000)
sigma2 = 0.05
mu = 17
ax.fill_between(x, 0, 0.5, edgecolor = 'r', hatch = 'xxx', facecolor = 'None')

y = (1/np.sqrt(2*np.pi*sigma2))*np.exp(-((x-mu)**2/2*sigma2))

ax.plot(x, y, color = 'b')
ax.fill_between(x,0,y, edgecolor = 'b', hatch = '////', facecolor = 'None', alpha = 0.6)

ax.set_xlim(1,40)
ax.set_ylim(0)
ax.set_ylabel('Frequency/probability density', fontsize = 18)
ax.set_xlabel('Parameter', fontsize = 18)
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(0.05))
ax.set_yticklabels([])
ax.tick_params(which = 'both', direction = 'in', top = True, right = True, labelsize = 14)
ax.tick_params(which = 'minor', length = 3)
ax.tick_params(which = 'major', length = 6)
ax.annotate('Uniform prior distribution', xy=(30, 0.5), xytext=(27,0.8), color = 'r', fontsize = 16, arrowprops=dict(facecolor = 'r', edgecolor = 'r'))
ax.annotate('Posterior distribution', xy=(13.5, 1.3), xytext=(2,1.4), color = 'b', fontsize = 16, arrowprops=dict(facecolor = 'b', edgecolor = 'b'))
plt.savefig('example_distribution_bayesian_inference.png', dpi=800)
```

```python
def _values_from_sobol(
        cls,
        paths: Sequence[str],
        dim_num_samples: int,
        *,
        bounds: dict[str, dict[str, str | Sequence[float]]] | None = None,
        scramble: bool = True,
        bits: int = None,
        optimization: Literal["random-cd", "lloyd"] | None = None,
        rng=None
) -> NDArray:
    
    from scipy.stats.qmc import Sobol, scale

    num_paths = len(paths)
    kwargs = dict(
        d = num_paths,
        scramble = scramble,
        bits = bits,
        optimization = optimization,
        rng = rng
    )

    bounds = bounds or {}

    scaling = np.asarray(
        [bounds.get(path, {}).get("scaling","linear") for path in paths]
    )

    all_extents = [bounds.get(path, {}).get("extent", [0,1]) for path in paths]

    extent = np.asarray(
        [
        np.log10(all_extents[i] if scaling[i] == "log" else all_extents[i] for i in range(len(scaling)))
        ]
    ).T

    try:
        sampler = Sobol(**kwargs)

    except:
        pass

    sobol_samples = scale(
        sampler.random_base2(m = dim_num_samples), l_bounds=extent[0], u_bounds = extent[1]
        )
    
    for i in range(len(scaling)):
        if scaling[i] == "log":
            sobol_samples[:,i] = 10**sobol_samples[:,i]

    return sobol_samples.T

@classmethod
def from_sobol(
    cls,
    paths: Sequence[str],
    dim_num_samples: int,
    *,
    bounds: dict[str, dict[str, str | Sequence[float]]] | None = None,
    scramble: bool = True,
    bits: int = None,
    optimization: Literal["random-cd", "lloyd"] | None = None,
    rng=None,
    nesting_order: int | float | None = None,
    label: str | int | None = None,
) -> Self:
    
    kwargs = {
        "paths": paths,
        "dim_num_samples": dim_num_samples,
        "scramble": scramble,
        "bits": bits,
        "optimization": optimization,
        "rng": rng,
        "bounds": bounds,
    }

    values = cls._values_from_sobol(**kwargs)
    assert values is not None
    obj = cls(
        paths=paths,
        values=values,
        nesting_order=nesting_order,
        label=label
    )
    obj._values_method = "from_sobol"
    obj._values_method_args = kwargs
    
    return obj
    

```
