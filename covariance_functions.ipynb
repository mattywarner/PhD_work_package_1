```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
r = np.linspace(0,5,1000)
l = 1
abs_exp_kernel = np.exp(-r/l)
RBF_kernel = np.exp(-r**2/(2*l**2))
mat_cov_nu_1_point_5 = (1+(np.sqrt(3)/l)*r)*np.exp(-(np.sqrt(3)/l)*r)
mat_cov_nu_2_point_5 = (1+(np.sqrt(5)/l)*r + (5/(3*l**2))*r**2)*np.exp(-(np.sqrt(5)/l)*r)
```

```python
plt.plot(r, abs_exp_kernel, label = "Absolute exponential kernel")
plt.plot(r, RBF_kernel, label = 'RBF kernel')
plt.plot(r, mat_cov_nu_1_point_5, label = r'Matern kernel, $\nu$ = 1.5')
plt.plot(r, mat_cov_nu_2_point_5, label = r'Matern kernel, $\nu$ = 2.5')
plt.xlabel('Input distance, r')
plt.ylabel('Covariance, k(r)')
plt.legend()
```
